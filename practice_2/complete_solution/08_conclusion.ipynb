# Predicting Product Sales
## 8. Conclusion & Future Directions

### Key Findings
- Gradient boosting regressors were successfully trained to predict transaction-level `Sales_Revenue`.
- The models achieved an $R^2$ score of approximately **0.485** (48.5% of variance explained), with the best model being **HistGradientBoosting** (Sklearn).
- **Why the score is around 48.5% instead of 100%:** To prevent trivial target leakage, we removed the `quantity` and `price` columns from our features, as `Sales_Revenue` is exactly `quantity * price`. Predicting the transaction revenue using only demographics (age, gender), shopping mall, payment method, and date features means the model must estimate the average revenue for a category without knowing how many items (1 to 5) the customer purchased. This demonstrates a realistic machine learning scenario where context explains a significant portion of, but not all, consumer spending variation.
- **HistGradientBoosting** and **LightGBM** models achieved the highest $R^2$ scores, while also offering rapid training and prediction times.

### Preprocessing Alignment and Next Steps
For further system optimization, the following adjustments are recommended:
1. **Native Categorical Features:** Rather than using One-Hot encoding, configure LightGBM and CatBoost to read raw categorical variables natively. This decreases representation dimensionality, accelerates model training, and avoids high-cardinality sparse arrays.
2. **Log Transformation:** To handle extreme outlier revenue transactions and balance the highly right-skewed target variable distribution, apply log-scaling during preprocessing:
   $$y_{\text{log}} = \log(y + 1)$$
   Predict values in log-scale and transform predictions back using exponential transformation.
3. **Age Generational Binning:** Continuous age values can be binned into cohorts (e.g., Generation Z, Millennials, Gen X, Boomers) to reflect purchasing behavior trends and simplify tree splitting.
