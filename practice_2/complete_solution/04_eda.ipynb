# Predicting Product Sales
## 4. Exploratory Data Analysis

We perform a detailed exploratory data analysis to understand data distributions, seasonal behaviors, and relations between demographics/locations and sales revenue.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

CLEANED_DATA_PATH = r'./data/preprocessed-data/preprocessed_customer_shopping_data.csv'
df = pd.read_csv(CLEANED_DATA_PATH, parse_dates=['invoice_date'])
print('Dataset shape:', df.shape)
df.head()
```

### 4.1 Target Variable (Sales_Revenue) Distribution
We inspect how sales revenue is distributed and compute its skewness.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Sales_Revenue'], bins=50, kde=True, color='teal')
plt.title('Distribution of Sales Revenue', fontsize=14, fontweight='bold')
plt.xlabel('Sales Revenue')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print(f'Skewness of Sales Revenue: {df["Sales_Revenue"].skew():.2f}')
```

### 4.2 Age and Gender Distributions
We analyze customer demographic distributions.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['age'], bins=20, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Age Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Age')

gender_counts = df['gender'].value_counts()
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
            colors=['lightcoral', 'skyblue'], startangle=90, wedgeprops=dict(edgecolor='black'))
axes[1].set_title('Gender Distribution', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 4.3 Transaction Distributions by Categorical Fields
We check transaction volumes across different categories, payment methods, and malls.

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 15))

sns.countplot(data=df, y='category', order=df['category'].value_counts().index, palette='viridis', ax=axes[0])
axes[0].set_title('Transactions by Category', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Transaction Count')

sns.countplot(data=df, x='payment_method', order=df['payment_method'].value_counts().index, palette='Set2', ax=axes[1])
axes[1].set_title('Transactions by Payment Method', fontsize=12, fontweight='bold')

sns.countplot(data=df, y='shopping_mall', order=df['shopping_mall'].value_counts().index, palette='coolwarm', ax=axes[2])
axes[2].set_title('Transactions by Shopping Mall', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Transaction Count')

plt.tight_layout()
plt.show()
```

### 4.4 Sales Revenue Analysis by Shopping Mall
We visualize how the distributions of sales revenue differ across different malls.

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Sales_Revenue', y='shopping_mall', palette='Set3')
plt.title('Sales Revenue Distribution across Shopping Malls', fontsize=14, fontweight='bold')
plt.xlabel('Sales Revenue')
plt.ylabel('Shopping Mall')
plt.tight_layout()
plt.show()
```

### 4.5 Temporal Shopping Behaviors
We analyze how transaction frequency varies across months and days of the week.

```python
df['YearMonth'] = df['invoice_date'].dt.to_period('M')
df['DayOfWeek'] = df['invoice_date'].dt.day_name()

fig, axes = plt.subplots(2, 1, figsize=(14, 12))

monthly_tx = df.groupby('YearMonth').size()
monthly_tx.plot(kind='line', marker='o', color='teal', linewidth=2, ax=axes[0])
axes[0].set_title('Monthly Transaction Trend (Jan 2021 - Feb 2023)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Transaction Count')

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(data=df, x='DayOfWeek', order=day_order, palette='magma', ax=axes[1])
axes[1].set_title('Transactions by Day of the Week', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 4.6 Weekly Customer Behavior per Shopping Mall

```python
plt.figure(figsize=(14, 6))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_mall = df.groupby(['DayOfWeek', 'shopping_mall']).size().reset_index(name='Transaction_Count')
sns.lineplot(data=daily_mall, x='DayOfWeek', y='Transaction_Count', hue='shopping_mall', marker='o', sort=False)
plt.title('Weekly Transaction Patterns by Shopping Mall', fontsize=14, fontweight='bold')
plt.ylabel('Transaction Count')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 4.7 Key Insights
1. **Stable Monthly Demand:** Customer transactions remain extremely stable month-over-month (ranging between 3,500 and 4,000 invoices per mall/month), showing no significant year-end holiday spikes.
2. **Uniform Weekday Volume:** Shopping frequencies are evenly distributed across all days of the week, with no major weekend surges.
3. **High Target Skewness:** `Sales_Revenue` is highly right-skewed. The distribution contains several distinct local peaks, corresponding to combinations of discrete product categories and quantities.
