# Predicting Product Sales
## 3. Data Cleaning

Based on findings from Step 2, we will perform the following operations:
1. Filter out the incomplete data of March 2023.
2. Create the target variable: `Sales_Revenue = quantity * price`.
3. Drop identifier columns and columns contributing to leakage (`invoice_no`, `customer_id`, `quantity`, `price`).
4. Save the cleaned dataset for EDA and model training.

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')

RAW_DATA_PATH = r'../data/raw-data/customer_shopping_data.csv'
df = pd.read_csv(RAW_DATA_PATH)
print('Original dataset shape:', df.shape)
```

### 3.1 Filter Out Incomplete Temporal Data
We filter out transactions recorded on or after March 1st, 2023, to ensure seasonal calculations and model benchmarking remain unbiased.

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
cleaned_df = df[df['invoice_date'] < '2023-03-01'].copy()
print('Shape after filtering dates:', cleaned_df.shape)
print('New maximum invoice date:', cleaned_df['invoice_date'].max())
```

### 3.2 Engineer the Target Variable
The prediction target is `Sales_Revenue`, computed as the product of `quantity` and `price`.

```python
cleaned_df['Sales_Revenue'] = cleaned_df['quantity'] * cleaned_df['price']
print(cleaned_df['Sales_Revenue'].describe())
```

### 3.3 Drop Non-Predictive and Leaky Columns
- `invoice_no` and `customer_id` are identifiers with no predictive power.
- `quantity` and `price` are dropped because their product is exactly the target variable. Keeping them would lead to trivial target leakage.

```python
cleaned_df = cleaned_df.drop(columns=['invoice_no', 'customer_id', 'quantity', 'price'])
print('Cleaned dataset columns:', list(cleaned_df.columns))
cleaned_df.head()
```

### 3.4 Save Cleaned Dataset

```python
out_dir = r'./data/preprocessed-data'
os.makedirs(out_dir, exist_ok=True)
out_path = os.path.join(out_dir, 'preprocessed_customer_shopping_data.csv')

cleaned_df.to_csv(out_path, index=False)
print('Cleaned dataset saved to:', out_path)
print('Final shape:', cleaned_df.shape)
```
