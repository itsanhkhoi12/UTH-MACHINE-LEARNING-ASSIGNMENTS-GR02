### Model Selection Rationale
We benchmark four modern Gradient Boosting variants:
- **HistGradientBoostingRegressor (Sklearn):** Native support for missing values, bin-based splitting.
- **XGBRegressor (XGBoost):** Highly optimized gradient boosting using tree-pruning and regularization.
- **LGBMRegressor (LightGBM):** Leaf-wise tree growth, fast training speed, low memory usage.
- **CatBoostRegressor (CatBoost):** Symmetric trees, robust categorical features handling, prevents overfitting.
