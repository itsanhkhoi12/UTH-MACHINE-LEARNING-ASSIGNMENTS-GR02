```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("./data/raw-data/customer_shopping_data.csv")
```

```python
data['invoice_date'] = pd.to_datetime(data['invoice_date'], format='%d/%m/%Y')

data = data[data['invoice_date'] < '2023-03-01'] 

# Tạo Biến mục tiêu
data['Sales_Revenue'] = data['quantity'] * data['price']

# Chống rò rỉ dữ liệu & Xóa cột định danh
data = data.drop(columns=['invoice_no', 'customer_id', 'quantity', 'price'])
```

```python
data.to_csv('./data/preprocessed-data/preprocessed_customer_shopping_data.csv',index=False)
```
