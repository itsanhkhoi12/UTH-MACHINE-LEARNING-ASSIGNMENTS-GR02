## 8.2 Các phát hiện chính

1. Email spam trung bình có độ dài lớn hơn, chứa nhiều chữ số, dấu chấm than và các từ mang tính quảng cáo/tiếp thị hơn email bình thường.
2. Các đặc trưng cụm 2 từ (Bigram TF-IDF) giúp cải thiện đáng kể khả năng nhận diện nhờ bắt được các cụm từ spam kinh điển như "click here" hay "free offer".
3. Hằng số làm mịn Laplace (`alpha`) ít nhạy cảm trong khoảng 0.001 - 2.0; tham số tốt nhất tìm được qua Cross-Validation là `alpha = 0.001` cho điểm F1 tối ưu.
4. Lớp mô hình tự viết đạt độ chính xác tương đương tuyệt đối với thư viện Scikit-learn, khẳng định thuật toán đã được cài đặt chính xác.
5. Phân phối lớp trong tập dữ liệu tương đối cân bằng (~49% ham / 51% spam), do đó không cần áp dụng các kỹ thuật lấy mẫu lại hoặc điều chỉnh trọng số lớp.


## 8.3 Tra cứu siêu tham số của mô hình

| Tham số | Kiểu dữ liệu | Mặc định | Tác động |
|-----------|------|---------|--------|
| `alpha` | float | 1.0 | Hằng số làm mịn Laplace; giá trị lớn hơn sẽ tăng mức độ chính quy hóa |
| `force_alpha` | bool | True | Nếu False, alpha sẽ được tự động làm tròn thành 1e-10 nếu thiết lập nhỏ hơn |
| `fit_prior` | bool | True | Nếu False, mô hình sử dụng phân phối lớp đều thay vì tính toán từ dữ liệu |
| `class_prior` | array hoặc None | None | Ghi đè các giá trị xác suất tiên nghiệm cố định cho từng lớp |


## 8.4 Hạn chế và Hướng phát triển

| Hạn chế | Hướng cải tiến đề xuất |
|------------|-----------------------|
| TF-IDF bỏ qua thứ tự từ | Thử nghiệm các dải n-gram rộng hơn hoặc nhúng từ theo vị trí |
| Không hiểu ngữ nghĩa sâu | Tinh chỉnh (fine-tune) một mô hình Transformer như DistilBERT |
| Từ vựng tĩnh | Áp dụng học tăng trưởng (incremental learning) hoặc học trực tuyến |
| Chỉ hỗ trợ tiếng Anh | Bổ sung thư viện tiền xử lý đa ngôn ngữ |
| Không tận dụng header email | Tích hợp thêm siêu dữ liệu về người gửi, định tuyến và thông tin reply-to |
